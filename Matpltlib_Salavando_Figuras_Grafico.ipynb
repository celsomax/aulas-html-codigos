Comparando gráfico de barras

```
import matplotlib.pyplot as plt
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]
z = [200, 25, 400, 330, 100]
titulo = "Matpltlib Grafico de dispersão - Scatterplo"
eixoX = "Eixo X"
eixoY = "Eixo Y"

plt.title(titulo)
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.scatter(x1, y1, label = "Meus pontos", color="r", marker=".", s=z)
plt.plot(x1, y1,  color="k", linestyle="-", linewidth=5)
plt.legend()
#plt.show()
plt.savefig("figura1.png", dpi=150)

```
